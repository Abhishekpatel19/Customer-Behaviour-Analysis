```python
df = pd.read_csv("customer_shopping_behavior.csv")
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns = {'purchase_amount_(usd)':'purchase_amount'})
df.columns
```

```python
# create columns
lables = ['young adult','adult','middle_age','senior']
df['age_group'] =pd.qcut(df['age'],q=4, labels=lables)
df.head()
```

```python
df[['age','age_group']].head(10)
```

```python
# create a purchase_frquency-days column
frequency_mapping = {
    'Fortnightly': 14,
    'Monthly': 30,
    'Quarterly': 90,
    'Yearly': 365,
    'Bi-weekly': 14,
    'Every 3 Months': 90,
    'Weekly' : 7,
    'Annually' : 365

}
df['purchase_frquency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frquency_days','frequency_of_purchases']].head(10)
```

```python
# two colomns discount applied and promo code used has same value as discount is given only when the promo code is used ...see two columns has same values
df[['discount_applied','promo_code_used']].head(10)
```

```python
# are all values same?
(df['discount_applied']== df['promo_code_used']).all()
```
